### 2.7 True or false

#### 1/ Is $2^{n+1} = O(2^n)$?

True.

$$ 0 \le f(n) = 2^{n+1} = 2 \times 2^n \lt 3\times2^n = 3g(n)$$

Exist a constant `c` such that $ 0 \le f(n) \le g(n) $. (1)
So, by defintion, we have $ 2^{n+1} = O(2^n) $


#### 2/ Is $ 2^{2n} = O(2^n)$?

False.
There's no `c` that can make (1) true.

### 2.8. For each of the following pairs of functions, either $f(n)$ is in $O(g(n))$, $f(n)$ is in $\Omega(g(n))$, or $f(n)=\Theta(g(n))$. Determine which relationship is correct and briefly explain why.

#### 1/ $ f(n) = \log(n^2)$; $ g(n) = \log(n) + 5 $

$$ f(n) = \log(n^2) = 2\log(n) \le 2(\log(n) + 5) = 2g(n) $$

And

$$ g(n) \le f(n), \forall n \ge n_0 = 2^5 $$

So, $ f(n) = \Theta(g(n)) $



#### 2/ $ f(n) = \sqrt(n)$; $ g(n) = \log(n^2) $






#### 3/ $ f(n) = \log^2(n)$; $ g(n) = \log(n) $

$$ \lim_{n \to \infty}\frac{f(n)}{g(n)} 
= \lim_{n \to \infty} \frac{\log^2(n)}{\log(n)}
= \lim_{n \to \infty} \log(n)
= \infty $$

So $ f(n) = \Omega(g(n)) $


#### 4/ $ f(n) = n$; $ g(n) = \log^2(n) $



#### 5/ $ f(n) = n\log(n) + n$; $ g(n) = \log(n) $


$$ \lim_{n \to \infty}\frac{f(n)}{g(n)}  $$
$$ = \lim_{n \to \infty} \frac{n\log(n) + n}{\log(n)} $$
$$ = \lim_{n \to \infty} \left(n + \frac{n}{\log(n)}\right) $$
$$ = \infty $$

So $ f(n) = \Omega(g(n)) $



#### 6/ $ f(n) = 10$; $ g(n) = \log(10) $


$$ 0 \le f(n) = 10 \le 10\log(10) = 10g(n) $$ 
And 
$$ 0 \le g(n) = \log(10) \le 10 = f(n) $$

So $ f(n) = \Theta(g(n)) $



#### 7/ $ f(n) = 2^n$; $ g(n) = 10n^2 $

Easy to see that $f(n)$ grows much faster than $g(n)$ with sufficient $n$. So $ f(n) = \Omega(g(n)) $



#### 8/ $ f(n) = 2^n$; $ g(n) = 3^n $


$$ f(n) = 2^n \le 3^n = g(n), \forall n \le 1 $$

So $ f(n) = O(g(n)) $
