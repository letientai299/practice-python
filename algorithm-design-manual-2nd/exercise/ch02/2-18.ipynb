### 2.18 List the functions below from the lowest to the highest order. If any two or more are of the same order, indicate which.

$$
\begin{array}{llll}
n              & 2^n        & n\log n & \ln n        \\
n - n^3 + 7n^5 & \log n     & \sqrt n & e^n          \\
n^2  + \ln n   & n^2        & 2^{n-1} &\log(\log n) \\
n^3             & (\log n)^2 & n!     & n^{1+e}\ \text{where}\ 0 \le e < 1
\end{array}
$$



The functions are listed from top (lowest) to bottom (highest). Functions that have same order is on the same row.

$

$

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,15))
ax = plt.subplot(111)

x = np.arange(0.1, 50, dtype=float);
def show(f, flabel, option=""):
    ax.semilogy(x, f(x), option, label=flabel)
    ax.legend()
    
show(lambda x: x, "$x$")
show(lambda x: x - x**3 + 7*x**5, "$x - x^3 + 7x^5$")
show(lambda x: x**2 + np.log(x), "$x^2 + \ln x$")
show(lambda x: x**3, "$x^3$")

show(lambda x: np.power(2, x), "$2^x$", '^')
show(lambda x: np.log2(x), "$log(x)$", '^')
show(lambda x: x**2, "$x^2$", '^')
show(lambda x: (np.log2(x))**2, "$(log(x))^2$", '^')

show(lambda x: x* np.log2(x), "$x\log(x)$", 's')
show(lambda x: np.sqrt(x), "$\sqrt{x}$", 's')
show(lambda x: np.power(2, x-1), "$2^{x-1}$", 's')
from scipy.misc import factorial
show(lambda x: factorial(x), "$x!$", 's')

show(lambda x: np.log(x)/np.log(np.e), "$\ln(x)$", 'o')
show(lambda x: np.exp(x), "$e^x$", 'o')
show(lambda x: np.log(np.log(x)), "$\log(\log(x))$", 'o')
show(lambda x: np.power(x, 1.11), "$x^{1 + 0.11}$", 'o')


ax.legend(loc='center left', bbox_to_anchor=(1, 0.3))
```
