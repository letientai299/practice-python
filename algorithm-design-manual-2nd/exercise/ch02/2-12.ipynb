```python
%matplotlib inline
import numpy as np
from matplotlib.pyplot import *

def show(f, g, c_constant):
    x = np.linspace(0, 40, 100)
    plot(x, cf(x), 'r', x, c_constant*cg(x), 'b')
    legend(['$f(x)$', '${}g(x)$'.format(c_constant)])
```

### 2.12 For each of the following pairs of functions $f(n)$ and $g(n)$, give an appropriate positive constant $c$ such that $f(n) \le c.g(n)$ for all $n \gt 1$.

#### (a) $f(n) = n^2 + n + 1, g(n) = 2n^3$

c = 

```python
show(lambda n: n**2 + 1 + n,
     lambda n: 2 * n**3,
     2)   
```

##### (b) $f(n) = n\sqrt n + n^2, g(n) = n^2$

$c = 2$

```python
show(lambda x: x**2 + x*np.sqrt(x), # f(x)
     lambda x: x**2,                # g(x)
     2)                             # c
```

##### (c) $f(n) = n^2 - n + 1, g(n) = \frac{n^2}{2} $

$c = 2$

```python
show(lambda n: n**2 + 1 - n,
     lambda n: n**2 /2,
     2)          
```
